```python
import pandas as pd
import numpy as np
```

```python
data = {
    'City': ['Colombo', 'Kandy','Galle','Nuwera Eliya','Jaffna'],
    'Date' : ['2026-05-01','2026-05-01','2026-05-01','2026-05-01',np.nan],
    'Humidity': [80,85,70,90,75]
}
df = pd.DataFrame(data)
print("---Raw Data ---")
print(df)

df['Temperature'] = df['Temperature'].str.replace('F','',regex=False).astype(float)

avg_temp = df['Temperature'].mean()
df['Temperature'] = df['Temperature'].fillna(avg_temp)

df['Data'] = df['Data'].fillna('2026-05-01')
print("\n--- Cleaned Data ---")
print(df)
```
